```python
import pandas as pd
import zipfile
import os

# Define the path to the downloaded zip file
zip_file_path = '/tmp/Data_Entry_2017.csv.zip'

# Define the directory where the CSV should be extracted
extract_path = '/drive/MyDrive/data/'

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Now, load the extracted CSV file into a pandas DataFrame
df = pd.read_csv(os.path.join(extract_path, 'Data_Entry_2017.csv'))

# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
# Save the DataFrame to a CSV file in Google Drive
output_file_path = os.path.join(extract_path, 'Data_Entry_2017.csv')
df.to_csv(output_file_path, index=False)

print(f"DataFrame successfully saved to {output_file_path}")
```
